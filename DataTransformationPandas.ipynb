```python
import pandas as pd
```

## Data Transformation Functions in Pandas(Feature Engineering)
1. Apply()
2. Map()
3. Assign()
4. Replace(old, new)

```python
df = pd.read_csv("raw_data.csv")
```

```python
print(df)
```

```python
#Feature Engineering
df2 = df.copy()
```

```python
df2["tax"] = df2["income"].apply(lambda x: "20%" if x>=50000 else "10%")
```

```python
df2
```

```python
#Let's convert Male = > M, Female => F, Unknown => U
gender_map = {"Male":"M", "Female":"F", "Unknown":"U"}
df2["gender"] = df2["gender"].map(gender_map)
```

```python
df2
```

```python
#Let's increase the income of everyone by 10X

df2.assign(new_income=df2["income"]*1.1)
```

```python
df2["country"] = df2["country"].replace("USA", "US")
```
