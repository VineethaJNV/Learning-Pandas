## Data Transformation Functions in Pandas(Feature Engineering)
1. Apply()
2. Map()
3. Assign()
4. Replace(old, new)

```python
df = pd.read_csv("raw_data.csv")
```

```python
print(df)
```

```python
#Feature Engineering
df2 = df.copy()
```

```python
df2["tax"] = df2["income"].apply(lambda x: "20%" if x>=50000 else "10%")
```

```python
df2
```

```python
#Let's convert Male = > M, Female => F, Unknown => U
gender_map = {"Male":"M", "Female":"F", "Unknown":"U"}
df2["gender"] = df2["gender"].map(gender_map)
```

```python
df2
```

```python
#Let's increase the income of everyone by 10X

df2 = df2.assign(new_income=df2["income"]*1.1)
```

```python
df2["country"] = df2["country"].replace("USA", "US")
```

```python
df2
```

```python
df.columns
df2.columns = ["Id", "Name", "Age", "Country", "Gender", "Income", "Tax", "New_Income"]
```

```python
df2
```

```python
df2 = df2.fillna(100)
sorted_df=df2.sort_values(["Income", "Age"])#If income is same, sorting will be according to age
sorted_df.sort_index()
sorted_df.reset_index()
```

```python
#Ranking
sorted_df["Ranking"] = sorted_df["Income"].rank(ascending=False, method="max")
```
