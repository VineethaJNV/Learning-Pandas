```python
import pandas as pd
```

```python
#Sries are homogenous

s1 = pd.Series([1, 2, 3, 4, 5, ], index=["one", "two", "three", "four", "five"])#dtype is int64
print(s1)


s2 = pd.Series([1, 2, 3, 4, 5, "six"], index=["one", "two", "three", "four", "five",6])#dtype is Object
print(s2)
```

```python
#Vectorization in Pandas

s1 = pd.Series([1, 2, 3, 4, 5])
s2 = pd.Series([10, 20, 30, 40, 50])
print(s1+s2)
```

```python
#Series values can bbe changed but size can't be changed
s1 = pd.Series([1, 2, 3, 4, 5])#dtype is int64
s1[0] = 100
print(s1)
changes_sq = s1.drop(0) #size is immutable so new series will be created in the memory
print(changes_sq)

```
