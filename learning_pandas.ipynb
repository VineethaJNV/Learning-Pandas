```python
import pandas as pd
```

2 main data structures if Pandas are
1. Series -> 1D lablled array
2. DataFrame -> 2D labelled array

Lable is a named index and it is not just position

```python
info = {
    "name" : ["Vedhansh", "vineetha", "vani"],
    "CGPA" : [1.0, 10, 9.0]
}

df = pd.DataFrame(info)
print(df)
```

We can import data from .xlsx and .csv and from our SQL table as well and Pandas will be able to visualize them in a tabular format like above

```python
py_list = [1, 2, 3, 4, 5]
pd_series = pd.Series(py_list)
print(pd_series)
print(type(pd_series))
print(pd_series[0])
print(pd_series[2])
print(pd_series[4])
```

```python
age_series = pd.Series([23, 21, 25, 25], index=["Varshitha", "vpr", "vinee", "veni"])
print(age_series)
print(age_series["Varshitha"]) #Values can be accessed through labels as well
print(age_series[1])#index based access
print(age_series.index)
```
