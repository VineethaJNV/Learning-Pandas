# Data concatenation => Stacking DataFrames on top of each other (rows) or side by side (columns)

```python
import pandas as pd
```

```python
df1 = pd.DataFrame({
    "id":[1, 2, 3],
    "name":["Adam", "eve","bob"]
})

df2 = pd.DataFrame({
    "id":[4, 5, 6],
    "name":["Charlie", "Fester","Gilian"]
})

pd.concat([df1, df2]) #Concatenation is done based on rows 
pd.concat([df1, df2], ignore_index=True) #Concatenation is done based on rows 
```

```python
pd.concat([df1, df2], axis=1) #Concatenation is done based on columns   
```
