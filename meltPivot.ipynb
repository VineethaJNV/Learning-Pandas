## Melt => Wide to Long format


## Pivot => Long to Wide format


```python
import pandas as pd
```

```python
df = pd.DataFrame({
    "country":["USA", "USA", "India", "India"],
    "year":[2020, 2021, 202, 2021],
    "sales":[100, 120, 90, 110],
    "profit":[20, 25, 18, 22]
})
print(df)
```

Let's convert the above data into long format while preserving the data

```python
melted_df = df.melt(
    id_vars=["country", "year"], #The columns that we want to keep
    value_vars=["sales","profit"], #The columns that we want to transform into a single one
    var_name="Metrics",
    value_name="value"
)

melted_df.pivot(
    index=["country", "year"], #Index should be unique, country alone is not unique so..
    columns="Metrics", #Create columns from the metric data
    values="value"
)
```

## Let's convert the melted data into original wide

```python
melted_df.pivot(
    index=["country", "year"], #Index should be unique, country alone is not unique so..
    columns="Metrics", #Create columns from the metric data
    values="value"
)
```
