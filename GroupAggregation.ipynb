```python
import pandas as pd
```

```python
df = pd.read_csv("changed.csv")
print(df)
df.columns
```

```python
df.groupby("country")["income"].max()
```

```python
df.groupby("country")["income"].min()
```

```python
df.groupby("country")["income"].mean()
```

```python
df.groupby("gender")["income"].mean()
```

```python
#Aggregation

df.groupby("country")["income"].agg(["mean", "min", "max"])
df.groupby("country")["income"].aggregate(["mean", "min", "max"])


#We can rename the column names to display
df.groupby("country")["income"].agg(avg_salary="mean", min_salary="min", max_salary="max")
```

```python
#Group by Coutry and apply aggregate functions on different columns
df.groupby("country").agg({
    "income":"mean",
    "age":"mean"
})

df.groupby("country").agg({
    "income":"max",
    "age":"mean"
})

#Whn aggregate funcions are applied, it is better to rename the column so that they make better sense
df.groupby("country").agg(
    max_salary=("income","max"),
    avg_age=("age","mean")
)
```
