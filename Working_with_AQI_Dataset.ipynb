```python
import pandas as pd
```

```python
#Let's work with AQI Dataset

#step 01 -> Load the .csv file
df = pd.read_csv("globalAirQuality.csv")
df.describe()
df.info()
```

```python
#Selecting the data
#Accessing the values of particular columns
df.columns
print(df['country'])
print(df[["city", "country"]])#Case-sensitive
```

## To access the rows in dataframe => Through Label & Index based - loc & iloc
### location and index location

```python
df.loc[0]
df.loc[900:1000] #start idx : end idx (inclusive)
```

```python
df.iloc[0]
df.iloc[0:2]#the ending idx in exclusive with iloc
```

### How do I access the individual cell values in the dataframe?

```python
#loc function can be used  for the above purpose
#cell -> row, col
print(df.loc[1457, ["city","aqi"]])
print(df.loc[0:100, ["city","aqi"]])
```

### iloc[] works with indexes, so even columns should be passed as their respective indexes
### It won't work with column names

```python
df.iloc[1000:1250, 2]
df.loc[0:2, ["country", "city", "latitude", "longitude"]]
df.iloc[0:3, 1:5]
```

### To acccess individual cell names, we can use at and iat

```python
print(df.at[0, "city"])
print(df.iat[0,2])
```

## The dataframe methods used above create only a view and not a copy, excatly as in numpy

## To prevent accidental changes in the original data, uses copy chaining
df["city"].copy(), create a copy and then make changes, this way original data will remain unchanged
